```python
fraud = df.groupby(['Order Country'])['Order Status'].apply(lambda x: x[x.str.contains("SUSPECTED_FRAUD")].count())
```

```python
fraud.describe()
```

```python
x = fraud.sort_values(ascending=False).head()
print("Top 5 Countries Having most suspected frauds:\n")
pd.DataFrame(x)
dataframe = pd.DataFrame(x)
dataframe.rename(columns = {'Order Status': 'Suspected Frauds'}, inplace = False)
```

```python
selection = df['Order Country'] != "India"
```

```python
india = df[~selection]
```

```python
pending = india.groupby(['Order City'])['Order Status'].apply(lambda x: x[x.str.contains("PENDING")].count())
print("List of all cities and there respective no. of Pending order in India:\n ")
data = pd.DataFrame(pending)
data.rename(columns = {'Order Status': 'Pending Orders'}, inplace = False)
```
